Descobrir os países top 10:

```python
#.sort_values = ordena os valores de uma coluna em ordem crescente ou decrescente
#ascending=False = ordem decrescente
top_10 = tabela_imigrantes.sort_values("Total", ascending=False).head(10)
top_10
```

```python
#grafico com o total de cada um dos 10 paises com seaborn
sns.barplot(data = top_10, x = top_10.index, y = "Total")
```

```python
#mudar as barras para horizontal
#orient h = horizontal
sns.barplot(data = top_10, y = top_10.index, x = "Total", orient="h")
```

```python
ax = sns.barplot(data = top_10, y = top_10.index, x = "Total", orient="h")
ax.set_title("Top 10 países com mais imigrantes para o Canadá\n1980-2013", fontsize=18, loc="left")
ax.set_xlabel("Número de Imigrantes", fontsize = 12)
ax.set_ylabel(" ")
plt.show()
```

```python
#criar figura:
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data = top_10, y = top_10.index, x = "Total", orient="h")
ax.set_title("Top 10 países com maior imigração para o Canadá\n1980-2013", fontsize=16, loc="left")
ax.set_xlabel("Número de Imigrantes", fontsize = 14)
ax.set_ylabel(" ")
plt.show()

```

```python
#função para paleta de cores:
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data = top_10, y = top_10.index, x = 'Total', orient= 'h', palette=palette, hue=top_10.index)
    ax.set_title("Top 10 países com maior imigração para o Canadá\n1980-2013", fontsize=16, loc="left")
    ax.set_xlabel("Número de Imigrantes", fontsize = 14)
    ax.set_ylabel(" ")
    plt.show()
```

Gráficos interativos com animação:

```python

#Primeiro é necessário trnasformar transformar 'ano' em int 

dados_brasil['ano'] = dados_brasil['ano'].astype(int)

#importar a biblioteca plotly
import plotly.graph_objs as go

#se cria uma figura vazia
fig = go.Figure()
#add uma linha e definir o modo como a linha vai aparecer
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definir o layout da figura
fig.update_layout(
    title =dict(
        text = '<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
     updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )], 
    width=1000, height=500 )
# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

fig.show()
```
